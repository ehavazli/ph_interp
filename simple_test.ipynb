```python
## Simple interploation test
```

```python
import numpy as np
import matplotlib.pyplot as plt

from dev import phase_interp as phint
```

```python
# Using the simple array below:
sample = np.array([
    [2, 3, 2, ],
    [2, 1, 2, ],
    [2, 3, 2, ],
])

# Setting PS pixels as all larger than 1
ps_msk = (sample > 1).astype('bool')

# Plot sample overview
fig = plt.figure(figsize=(10, 5)) # New figure
ax = fig.add_subplot(1,2,1) # subplot
cax=ax.imshow(sample, cmap='viridis', vmin=1, vmax=3, interpolation='nearest')
ax.set_title(r'ifg')
cbar = fig.colorbar(cax, orientation='horizontal')
ax = fig.add_subplot(1,2,2) # subplot
cax=ax.imshow(np.ma.masked_where(sample*ps_msk==0, sample*ps_msk), 
              cmap='viridis', vmin=1, vmax=3, interpolation='nearest')
ax.set_title(r'Ps pixels')
cbar = fig.colorbar(cax, orientation='horizontal')
plt.show()
```

```python
# tests interpolation in this cell

ifg_re = np.copy(sample*ps_msk) # array to be resampled
# run interpolation
ifg_re = phint.chen_interp(ifg_re, radius=2, fill_value=0.0) 

print('Interpolated value result = ', ifg_re[1,2])

# Plot inerpolation overview
fig = plt.figure(figsize=(10, 5)) # New figure
ax = fig.add_subplot(1,3,1) # subplot
cax=ax.imshow(sample, cmap='viridis', vmin=1, vmax=3, interpolation='nearest')
ax.set_title(r'ifg orig')
cbar = fig.colorbar(cax, orientation='horizontal')
ax = fig.add_subplot(1,3,2) # subplot
cax=ax.imshow(ps_msk, cmap='bwr', interpolation='nearest')
ax.set_title(r'Ps mask')
cbar = fig.colorbar(cax, orientation='horizontal')
ax = fig.add_subplot(1,3,3) # subplot
cax=ax.imshow(ifg_re, cmap='viridis', vmin=1, vmax=3, interpolation='nearest')
ax.set_title(r'ifg interp')
cbar = fig.colorbar(cax, orientation='horizontal')
plt.show()
```
